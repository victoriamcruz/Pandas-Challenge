```python
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"



# Read Purchasing File and store into Pandas data frame
purchase_data_df = pd.read_csv(file_to_load)

player_demographics = purchase_data_df.loc[:, ['Gender','SN', 'Age']]

purchase_data_df.head()
```

## Player Count

* Display the total number of players


```python
# Find unique players to calculate the total number of players
unique_names = purchase_data_df["SN"].unique()

total = len(unique_names)

# Create a data frame to display total number of players
total_players = pd.DataFrame({"Total Players": [total]})
total_players
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
#identify variables
unique_items = purchase_data_df["Item ID"].value_counts()
average_price = purchase_data_df["Price"].mean()
number_of_purchase = purchase_data_df["Purchase ID"].value_counts()
total_revenue = purchase_data_df["Price"].sum()

raw_data_info = ({
    "Number of Unique Items": [len(unique_items)],
             "Average Price": [average_price],
       "Number of Purchases": [len(number_of_purchase)],
             "Total Revenue": [total_revenue]
})

#formatting for amounts 
pd.options.display.float_format = '${:,.2f}'.format

info_df = pd.DataFrame(raw_data_info, columns=["Number of Unique Items", "Average Price", "Number of Purchases", "Total Revenue"])

info_df
```

## Gender Demographics

```python
Gender_df = purchase_data_df[["SN", "Gender"]]

Gender_df = Gender_df.drop_duplicates(subset=['SN'])

gender_total = Gender_df["Gender"].value_counts()

sometotalvalue = Gender_df.drop_duplicates(subset=['SN'])

percent_gender = 100. * Gender_df['Gender'].value_counts() / len(sometotalvalue)

df = pd.DataFrame({'Total Count': gender_total, 'Percentage of Players': percent_gender})  

#formatting for percentages
pd.options.display.float_format = '{:,.2f}%'.format

df
```

## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender

* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
total_purchase_value = purchase_data_df.groupby(["Gender"]).count()["Price"]

Gender_df = purchase_data_df[["SN", "Gender"]]

Purchase_count = Gender_df["Gender"].value_counts()

Total_purchase = purchase_data_df.groupby(["Gender"]).sum()["Price"]

average_price = Total_purchase / Purchase_count

Avg_Purchase_per_Person = Total_purchase / gender_total

pd.options.display.float_format = '${:,.2f}'.format

df = pd.DataFrame ({
                     'Purchase Count' : Purchase_count,
                     'Average Purchase Price': Total_purchase,
                     'Total Purchase Value Purchases': average_price,
                     'Avg Total Purchase per Person': Avg_Purchase_per_Person
})
df
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
bins = [0, 9, 14, 19, 24, 29, 34, 39, 100]

group_names = ['<10', '10-14', '15-19', '20-24', '25-29', '30-34', '35-39', '40+']

player_demographics["Age Group"] = pd.cut(x = player_demographics['Age'], bins = bins, labels = group_names, include_lowest = False)

player_demographics
```

```python
player_demographics = player_demographics.drop_duplicates()

age_range = player_demographics.groupby(['Age Group']).count()['Age']

total_age = age_range / len(player_demographics) * 100

age_demographics = pd.DataFrame({"Total Count": age_range, "Percentage of Players": total_age})

pd.options.display.float_format = '{:,.2f}%'.format

age_demographics
```

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
bins = [0, 9, 14, 19, 24, 29, 34, 39, 100]

group_names = ['<10', '10-14', '15-19', '20-24', '25-29', '30-34', '35-39', '40+']

purchase_data_df["Age Group"] = pd.cut(x = purchase_data_df['Age'], bins = bins, labels = group_names, include_lowest = False)
purchase_count = purchase_data_df.groupby(['Age Group']).count()['Age']

total_purchase_value = purchase_data_df.groupby(['Age Group']).sum()['Price']

avg_purchase_price = total_purchase_value / purchase_count

player_demographics = player_demographics.drop_duplicates()

age_range = player_demographics.groupby(['Age Group']).count()['Age']

avg_per_person = total_purchase_value / age_range

pd.options.display.float_format = '${:,.2f}'.format

purchasing_analysis_age = pd.DataFrame({"Purchase Count": purchase_count, "Average Purchase Price": avg_purchase_price, "Total Purchase Value": total_purchase_value, "Avg Total Purchase per Person": avg_per_person})

purchasing_analysis_age
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
purchase_count_sn = purchase_data_df.groupby(['SN']).count()['Price']

total_purchase_sn = purchase_data_df.groupby(['SN']).sum()['Price']

avg_purchase_price_sn = total_purchase_sn / purchase_count_sn

top_spenders = pd.DataFrame({"Purchase Count": purchase_count_sn, "Average Purchase Price": avg_purchase_price_sn, "Total Purchase Value": total_purchase_sn})

top_spenders.sort_values(by='Total Purchase Value', ascending=False)
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
popular_items = purchase_data_df.loc[:, ['Item ID','Item Name', 'Price']]

pop_items_total = purchase_data_df.groupby(['Item ID', 'Item Name']).count()['Price']

pop_items_sum = purchase_data_df.groupby(['Item ID', 'Item Name']).sum()['Price']

item_price = pop_items_sum / pop_items_total

most_popular_items = pd.DataFrame({"Purchase Count": pop_items_total, "Item Price": item_price, "Total Purchase Value": pop_items_sum})

most_popular_items.sort_values(by='Purchase Count', ascending=False)
```
